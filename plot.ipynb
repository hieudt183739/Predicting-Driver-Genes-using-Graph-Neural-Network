```python
import pandas as pd
import numpy as np

gene = pd.read_csv('result/model/gene.txt', sep=',', header=None)
```

```python
from sklearn.datasets import make_blobs
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt

path = "./data/luad/"     # specific cancer,e.g.luad
pos = pd.read_csv('data/luad/pos.txt', header=None)
neg = pd.read_csv('data/luad/neg.txt', header=None)
index = pd.concat([pos,neg]).reset_index(drop=True).to_numpy().reshape(-1)   # index gene co nhan trong file gene_names.txt
label = np.zeros(2366)
label[0:179] = 1
```

```python
tsne = TSNE(n_components=2, random_state=42)
X_tsne = tsne.fit_transform(gene.iloc[index,:])
```

```python
plt.figure(figsize=(8, 6))
scatter = plt.scatter(X_tsne[:, 0], X_tsne[:, 1], c=label ,cmap='viridis', marker='*')
plt.colorbar(scatter, ticks=[0, 1, 2])
plt.title('t-SNE Visualization')
plt.xlabel('t-SNE Component 1')
plt.ylabel('t-SNE Component 2')
plt.show()
```

```python
gene_before = pd.read_csv('data/luad/P.feat-final.csv', header=None, index_col=0)
tsne = TSNE(n_components=2, random_state=42)
X_tsne = tsne.fit_transform(gene_before.iloc[index, :])
```
